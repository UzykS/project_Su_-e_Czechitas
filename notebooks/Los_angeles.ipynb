```python
crime_categories = set(crime_dict.values())

# Вивести унікальні категорії
print("Унікальні категорії злочинів:")
for category in crime_categories:
    print(category)

# Порахувати кількість унікальних категорій
кількість_категорій = len(crime_categories)
print(f"\nЗагальна кількість унікальних категорій: {кількість_категорій}")
```

```python
merge_categories = {
    'Criminal Misconduct or Miscellaneous Offenses': 'Miscellaneous Offenses',
    'Theft and Property Offenses or Fraud and Deception': 'Theft and Property Offenses'
}

# Створення нового словника зі зміненими категоріями
merged_crime_dict = {crime: merge_categories.get(category, category) for crime, category in crime_dict.items()}

# Вивід зміненого словника
print("Змінений словник:")
for crime, category in merged_crime_dict.items():
    print(f"{crime}: {category}")
```

```python
los_angeles['Crime'] = los_angeles['Crm Cd Desc'].map(crime_dict)
```
