```python
crime_dict = {
    'Other Assaults': 'Assault and Violent Crimes',
    'All Other Offenses': 'Miscellaneous Offenses',
    'Weapon Violations': 'Firearm and Weapon-Related Crimes',
    'Thefts': 'Theft and Property Offenses',
    'Burglary Non-Residential': 'Theft and Property Offenses',
    'Aggravated Assault Firearm': 'Assault and Violent Crimes',
    'Theft from Vehicle': 'Theft and Property Offenses',
    'Disorderly Conduct': 'Miscellaneous Offenses',
    'Vandalism/Criminal Mischief': 'Miscellaneous Offenses',
    'Arson': 'Miscellaneous Offenses',
    'Fraud': 'Fraud and Deception',
    'Robbery No Firearm': 'Theft and Property Offenses',
    'Vagrancy/Loitering': 'Miscellaneous Offenses',
    'Motor Vehicle Theft': 'Theft and Property Offenses',
    'Recovered Stolen Motor Vehicle': 'Theft and Property Offenses',
    'Robbery Firearm': 'Theft and Property Offenses',
    'Embezzlement': 'Fraud and Deception',
    'Rape': 'Sexual Offenses',
    'DRIVING UNDER THE INFLUENCE': 'Traffic and Vehicle-Related Offenses',
    'Forgery and Counterfeiting': 'Fraud and Deception',
    'Narcotic / Drug Law Violations': 'Drug-Related Offenses',
    'Burglary Residential': 'Theft and Property Offenses',
    'Other Sex Offenses (Not Commercialized)': 'Sexual Offenses',
    'Liquor Law Violations': 'Miscellaneous Offenses',
    'Aggravated Assault No Firearm': 'Assault and Violent Crimes',
    'Homicide - Criminal': 'Assault and Violent Crimes',
    'Gambling Violations': 'Miscellaneous Offenses',
    'Prostitution and Commercialized Vice': 'Miscellaneous Offenses',
    'Public Drunkenness': 'Miscellaneous Offenses',
    'Receiving Stolen Property': 'Theft and Property Offenses',
    'Homicide - Gross Negligence': 'Assault and Violent Crimes',
    'Offenses Against Family and Children': 'Child-Related Offenses',
    'Homicide - Justifiable': 'Assault and Violent Crimes'
}
```

```python
denver.head(20)
```

```python
crime_categories = set(crime_dict.values())

# Вивести унікальні категорії
print("Унікальні категорії злочинів:")
for category in crime_categories:
    print(category)

# Порахувати кількість унікальних категорій
кількість_категорій = len(crime_categories)
print(f"\nЗагальна кількість унікальних категорій: {кількість_категорій}")
```

```python
denver['Crime'] = denver['Text_General_Code'].map(crime_dict)
```

```python
denver.rename(columns={'Text_General_Code' : 'Specific crime' , 'Dispatch_Date' : 'Date'}, inplace=True)
```

```python
denver['City'] = 'Denver'
```

```python
denver['Date'] = pd.to_datetime(denver['Date'], format='%Y-%m-%d')
denver['Date'] = denver['Date'].dt.strftime('%d/%m/%Y')
```

```python
denver = denver[["Date", "City", "Crime", "Specific crime"]]
```
