```python
b15 = pd.read_csv('../data_sources/boston/BOSTON2015.csv', encoding="utf-8", low_memory=False, usecols=['OCCURRED_ON_DATE', 'OFFENSE_CODE_GROUP', 'OFFENSE_CODE', 'SHOOTING'])
b16 = pd.read_csv('../data_sources/boston/BOSTON2016.csv', encoding="utf-8", low_memory=False, usecols=['OCCURRED_ON_DATE', 'OFFENSE_CODE_GROUP', 'OFFENSE_CODE', 'SHOOTING'])
b17 = pd.read_csv('../data_sources/boston/BOSTON2017.csv', encoding="utf-8", low_memory=False, usecols=['OCCURRED_ON_DATE', 'OFFENSE_CODE_GROUP', 'OFFENSE_CODE', 'SHOOTING'])
b18 = pd.read_csv('../data_sources/boston/BOSTON2018.csv', encoding="utf-8", low_memory=False, usecols=['OCCURRED_ON_DATE', 'OFFENSE_CODE_GROUP', 'OFFENSE_CODE', 'SHOOTING'])
b19 = pd.read_csv('../data_sources/boston/BOSTON2019.csv', encoding="utf-8", low_memory=False, usecols=['OCCURRED_ON_DATE', 'OFFENSE_CODE', 'SHOOTING'])
b20 = pd.read_csv('../data_sources/boston/BOSTON2020.csv', encoding="utf-8", low_memory=False, usecols=['OCCURRED_ON_DATE', 'OFFENSE_CODE', 'SHOOTING'])
b21 = pd.read_csv('../data_sources/boston/BOSTON2021.csv', encoding="utf-8", low_memory=False, usecols=['OCCURRED_ON_DATE', 'OFFENSE_CODE', 'SHOOTING'])
b22 = pd.read_csv('../data_sources/boston/BOSTON2022.csv', encoding="utf-8", low_memory=False, usecols=['OCCURRED_ON_DATE', 'OFFENSE_CODE', 'SHOOTING'])
```

```python
boston_old = pd.concat([b15, b16, b17, b18])
boston_new = pd.concat([b19, b20, b21, b22])
```

```python

# Create a dictionary with offense code as key and offense name as value
offenses_dict = dict(zip(boston_old['OFFENSE_CODE'], boston_old['OFFENSE_CODE_GROUP']))

# Convert the dictionary to a DataFrame
offenses_df = pd.DataFrame(list(offenses_dict.items()), columns=['OFFENSE_CODE', 'OFFENSE_DESCRIPTION'])

# Save the DataFrame to a CSV file
offenses_df.to_csv('../data_standartized/offences_dict.csv', index=False)
```

```python
boston_old["Specific crime"] = boston_old["OFFENSE_CODE"].map(offenses_dict)
```

```python
boston_new["Specific crime"] = boston_new["OFFENSE_CODE"].map(offenses_dict)
```

```python

boston = pd.concat([boston_old, boston_new])
```
